```python
import pandas as pd

criminalPlaceTop10 = pd.read_csv('./data/criminalPlaceTop10.csv')
criminalPlaceTop10
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

fig, ax = plt.subplots(figsize=(15,10))
bar_width = 0.15
index = np.arange(len(list(criminalPlaceTop10.columns)[1:]))

b1 = plt.bar(index - bar_width, list(criminalPlaceTop10.loc[index[0]])[1:], width=bar_width, alpha=0.4, color='red', label=criminalPlaceTop10.loc[index[0], ['범죄중분류']][0])
b2 = plt.bar(index, list(criminalPlaceTop10.loc[index[1]])[1:], width=bar_width, alpha=0.4, color='orange', label=criminalPlaceTop10.loc[index[1], ['범죄중분류']][0])
b3 = plt.bar(index + bar_width, list(criminalPlaceTop10.loc[index[2]])[1:], width=bar_width, alpha=0.4, color='yellow', label=criminalPlaceTop10.loc[index[2], ['범죄중분류']][0])
b4 = plt.bar(index + 2 * bar_width, list(criminalPlaceTop10.loc[index[3]])[1:], width=bar_width, alpha=0.4, color='green', label=criminalPlaceTop10.loc[index[3], ['범죄중분류']][0])

plt.xticks(index, list(criminalPlaceTop10.columns)[1:])

plt.xlabel('장소', size = 13)
plt.ylabel('범죄 건수', size = 13)
plt.legend()
plt.show()
```
