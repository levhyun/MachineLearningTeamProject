### 데이터 전처리 및 가공 - 03

##### 기본 전처리된 csv 파일 읽기

```python
import pandas as pd

criminalArea = pd.read_csv('./data/criminalArea.csv')
criminalArea
```

##### 필요없는 열 삭제하기

```python
criminalArea.drop('기타도시', axis='columns', inplace=True)
criminalArea.drop('도시이외', axis='columns', inplace=True)
print(criminalArea.columns)
```

##### 지역별 범죄 건수를 기준으로 정렬하기

```python
area = criminalArea[list(criminalArea.columns)[1:]].sum(axis=0)
area.sort_values(ascending=False, inplace=True)
area
```

##### 정렬된 지역별 범죄 건수 인덱스 생성하기

```python
areaColumns = list(criminalArea.columns)[:1] + list(area.index)
areaColumns
```

##### 지역별 범죄 건수를 기준으로 필터링하기

```python
criminalArea2 = criminalArea[areaColumns]
criminalArea2
```

##### 범죄별 건수를 기준으로 정렬하기

```python
criminal = criminalArea2.sum(axis=1)
criminal.sort_values(ascending=False, inplace=True)
criminal
```

##### 정렬된 범죄별 건수 인덱스 생성하기

```python
criminalIndex = list(criminal.index)
criminalIndex
```

##### 범죄별 건수를 기준으로 필터링하기

```python
finalCriminalArea = criminalArea2.iloc[criminalIndex].reset_index(drop=True)
finalCriminalArea
```

##### csv파일로 저장하기

```python
finalCriminalArea.to_csv('../DataVisualization/organizedData/criminalArea.csv', index=False)
```
