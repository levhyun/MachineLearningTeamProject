### 데이터 전처리 및 가공 - 01

##### 기본 전처리된 csv 파일 읽기

```python
import pandas as pd

criminalTimeAndDay = pd.read_csv('./data/criminalTimeAndDay.csv')
criminalTimeAndDay
```

##### 시간열 추출하기

```python
criminalTime = criminalTimeAndDay[list(criminalTimeAndDay.columns)[:9]]
criminalTime
```

##### 시간 포맷팅하기

```python
criminalTime.columns = list(column if column == '범죄중분류' else column.replace('-', ' ~ ').replace('시', ':').replace('분', '') for column in list(criminalTime.columns))
criminalTime.columns
```

##### 시간별 범죄 건수를 기준으로 정렬하기

```python
time = criminalTime[list(criminalTime.columns)[1:]].sum(axis=0)
time.sort_values(ascending=False, inplace=True)
time
```

##### 정렬된 시간별 범죄 건수 인덱스 생성하기

```python
timeColumns = list(criminalTime.columns)[:1] + list(time.index)
timeColumns
```

##### 시간별 범죄 건수를 기준으로 필터링하기

```python
criminalTime2 = criminalTime[timeColumns]
criminalTime2
```

##### 범죄별 건수를 기준으로 정렬하기

```python
criminal = criminalTime2.sum(axis=1)
criminal.sort_values(ascending=False, inplace=True)
criminal
```

##### 정렬된 범죄별 건수 인덱스 생성하기

```python
criminalIndex = list(criminal.index)
criminalIndex
```

##### 범죄별 건수를 기준으로 필터링하기

```python
finalCriminalTime = criminalTime2.iloc[criminalIndex].reset_index(drop=True)
finalCriminalTime
```

##### csv파일로 저장하기

```python
finalCriminalTime.to_csv('../DataVisualization/data/criminalTime.csv', index=False)
```

##### 요일열 추출하기

```python
criminalDay = criminalTimeAndDay[list(criminalTimeAndDay.columns)[:1] + list(criminalTimeAndDay.columns)[10:]]
criminalDay
```

##### 요일 포맷팅하기

```python
criminalDay.columns = list(column if column == '범죄중분류' else column + "요일" for column in list(criminalDay.columns))
criminalDay.columns
```

##### 요일별 범죄 건수를 기준으로 정렬하기

```python
day = criminalDay[list(criminalDay.columns)[1:]].sum(axis=0)
day.sort_values(ascending=False, inplace=True)
day
```

##### 정렬된 요일별 범죄 건수 인덱스 생성하기

```python
DayColumns = list(criminalDay.columns)[:1] + list(day.index)
DayColumns
```

##### 요일별 범죄 건수를 기준으로 필터링하기

```python
criminalDay2 = criminalDay[DayColumns]
criminalDay2
```

##### 범죄별 건수를 기준으로 정렬하기

```python
criminal = criminalDay2.sum(axis=1)
criminal.sort_values(ascending=False, inplace=True)
criminal
```

##### 정렬된 범죄별 건수 인덱스 생성하기

```python
criminalIndex = list(criminal.index)
criminalIndex
```

##### 범죄별 건수를 기준으로 필터링하기

```python
finalCriminalDay = criminalDay2.iloc[criminalIndex].reset_index(drop=True)
finalCriminalDay
```

##### csv파일로 저장하기

```python
finalCriminalDay.to_csv('../DataVisualization/data/criminalDay.csv', index=False)
```
