### 데이터 전처리 및 가공 - 02

##### 기본 전처리된 csv 파일 읽기

```python
import pandas as pd

criminalPlace = pd.read_csv('./data/criminalPlace.csv')
criminalPlace
```

##### 필요없는 열 삭제하기

```python
criminalPlace.drop('기타', axis='columns', inplace=True)
criminalPlace.drop('기타교통수단내', axis='columns', inplace=True)
print(criminalPlace.columns)
```

##### 이상한 데이터 변경하기

```python
criminalPlace.rename(columns={'아파트_연립다세대':'아파트'}, inplace=True)
criminalPlace.rename(columns={'유흥접객업소':'유흥업소'}, inplace=True)
criminalPlace.rename(columns={'숙박업소_목욕탕':'숙박업소 및 목욕탕'}, inplace=True)
```

##### 장소별 범죄 건수를 기준으로 정렬하기

```python
place = criminalPlace[list(criminalPlace.columns)[1:]].sum(axis=0)
place.sort_values(ascending=False, inplace=True)
place
```

##### 정렬된 장소별 범죄 건수 인덱스 생성하기

```python
placeColumns = list(criminalPlace.columns)[:1] + list(place.index)
placeColumns
```

##### 장소별 범죄 건수를 기준으로 필터링하기

```python
criminalPlace2 = criminalPlace[placeColumns]
criminalPlace2
```

##### 범죄별 건수를 기준으로 정렬하기

```python
criminal = criminalPlace2.sum(axis=1)
criminal.sort_values(ascending=False, inplace=True)
criminal
```

##### 정렬된 범죄별 건수 인덱스 생성하기

```python
criminalIndex = list(criminal.index)
criminalIndex
```

##### 범죄별 건수를 기준으로 필터링하기

```python
finalCriminalPlace = criminalPlace2.iloc[criminalIndex].reset_index(drop=True)
finalCriminalPlace
```

##### csv파일로 저장하기

```python
finalCriminalPlace.to_csv('../DataVisualization/organizedData/criminalPlace.csv', index=False)
```
